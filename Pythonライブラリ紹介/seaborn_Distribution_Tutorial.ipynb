# seaborn - Visualizing the distribution of a dataset  

[Visualizing the distribution of a dataset](https://seaborn.pydata.org/tutorial/distributions.html)  

`seaborn.distplot()`  
[seaborn.distplot](https://seaborn.pydata.org/generated/seaborn.distplot.html#seaborn.distplot)  

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
sns.set(color_codes=True)
```

## Plotting univariate distributions  

```python
x = np.random.normal(size=100)
sns.distplot(x);
```

```python
sns.distplot(x, kde=False, rug=True);
```

```python
sns.distplot(x, bins=20, kde=False, rug=True);
```

```python
sns.distplot(x, hist=False, rug=True);
```

```python
x = np.random.normal(0, 1, size=30)
bandwidth = 1.06 * x.std() * x.size ** (-1 / 5.)
support = np.linspace(-4, 4, 200)

kernels = []
for x_i in x:

    kernel = stats.norm(x_i, bandwidth).pdf(support)
    kernels.append(kernel)
    plt.plot(support, kernel, color="r")

sns.rugplot(x, color=".2", linewidth=3);
```

```python
from scipy.integrate import trapz
density = np.sum(kernels, axis=0)
density /= trapz(density, support)
plt.plot(support, density);
```

```python
sns.kdeplot(x, shade=True);
```

```python
sns.kdeplot(x)
sns.kdeplot(x, bw=.2, label="bw: 0.2")
sns.kdeplot(x, bw=2, label="bw: 2")
plt.legend();
```

```python
sns.kdeplot(x, shade=True, cut=0)
sns.rugplot(x);
```

### Fitting parametric distributions  

```python
x = np.random.gamma(6, size=200)
sns.distplot(x, kde=False, fit=stats.gamma);
```
